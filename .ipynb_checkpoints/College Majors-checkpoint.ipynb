The American Community Survey is a survey run by the US Census Bureau that collects data on everything from the affordability of housing to employment rates for different industries.

Here's a quick overview of the files we'll be working with:

majors-list.csv:
- List of majors with their FOD1P codes and major categories.
- Major categories are from Carnevale et al, "What's It Worth?: The Economic Value of College Majors." Georgetown University Center on Education and the Workforce, 2011. http://cew.georgetown.edu/whatsitworth

Three main data files:
- all-ages.csv
- recent-grads.csv (ages <28)
- grad-students.csv (ages 25+)
All contain basic earnings and labor force information. 
recent-grads contains more detailed breakdown, including by sex and by the type of job they got. Full headers below.
grad-students contains details on graduate school attendees.

Additionally, women-stem.csv contains data for scatter plot in associated DataLab post on women in science/technology jobs. It is a subset of recent-grads.csv. (Small easter egg: Check out my related Shiny app: https://bencasselman.shinyapps.io/new-test/)

Headers for  recent-grads.csv

Header | Description
---|---------
`Rank` | Rank by median earnings
`Major_code` | Major code, FO1DP in ACS PUMS
`Major` | Major description
`Major_category` | Category of major from Carnevale et al
`Total` | Total number of people with major
`Sample_size` | Sample size (unweighted) of full-time, year-round ONLY (used for earnings)
`Men` | Male graduates
`Women` | Female graduates
`ShareWomen` | Women as share of total
`Employed` | Number employed (ESR == 1 or 2)
`Full_time` | Employed 35 hours or more
`Part_time` | Employed less than 35 hours
`Full_time_year_round` | Employed at least 50 weeks (WKW == 1) and at least 35 hours (WKHP >= 35)
`Unemployed` | Number unemployed (ESR == 3)
`Unemployment_rate` | Unemployed / (Unemployed + Employed)
`Median` | Median earnings of full-time, year-round workers
`P25th` | 25th percentile of earnigns
`P75th` | 75th percentile of earnings
`College_jobs` | Number with job requiring a college degree
`Non_college_jobs` | Number with job not requiring a college degree
`Low_wage_jobs` | Number in low-wage service jobs

#### Get the total number of people in each Major_category for both datasets

```python
import pandas as pd

# read files
all_ages = pd.read_csv("all-ages.csv")
recent_grads = pd.read_csv("recent-grads.csv")

# instantiate dictionaries
all_ages_major_categories = dict()
recent_grads_major_categories = dict()

print(type(all_ages))

# summarize using pivot tables
#all_ages_pt = all_ages.pivot_table(index="Major_category", values="Total", aggfunc=np.sum)
#print(all_ages_pt)
#print(type(all_ages_pt))

for a in all_ages['Major_category'].value_counts().index.tolist():
    cur = all_ages["Major_category"] == a
    rows = all_ages.loc[cur]
    total = rows["Total"]
    all_ages_major_categories[a] = total.sum()

print(all_ages_major_categories)


for a in recent_grads['Major_category'].value_counts().index.tolist():
    cur = recent_grads["Major_category"] == a
    rows = recent_grads.loc[cur]
    total = rows["Total"]
    recent_grads_major_categories[a] = total.sum()

print(recent_grads_major_categories)
        
```

#### Use the Low_wage_jobs and Total columns to calculate the proportion of recent college graduates that worked low wage jobs

```python
low_wage_percent = recent_grads["Low_wage_jobs"].sum()/recent_grads["Total"].sum()
print(low_wage_percent)
```

#### Calculate the number of majors where recent grads did better than the overall population

##### Solution 1

```python
# All majors, common to both DataFrames
majors = recent_grads['Major'].value_counts().index

# sort values by Major
recent_grads.sort_values("Major",inplace=True)
all_ages.sort_values("Major",inplace=True)

# being sorted and knowing in advance the one-to-one match by Major, make a one-line comparison
recent_grads_lower_unemp_count = (recent_grads["Unemployment_rate"] < all_ages["Unemployment_rate"]).sum()
all_ages_lower_unemp_count = (recent_grads["Unemployment_rate"] > all_ages["Unemployment_rate"]).sum()

print(recent_grads_lower_unemp_count)
print(all_ages_lower_unemp_count)
```

##### Solution 2

```python
# All majors, common to both DataFrames
majors = recent_grads['Major'].value_counts().index

recent_grads_lower_unemp_count = 0
all_ages_lower_unemp_count = 0
for m in majors:
    recent_grads_row =  recent_grads[recent_grads['Major'] == m]
    all_ages_row = all_ages[all_ages['Major'] == m]
    
    recent_grads_unemp_rate = recent_grads_row['Unemployment_rate'].values[0]
    all_ages_unemp_rate = all_ages_row['Unemployment_rate'].values[0]
    
    if recent_grads_unemp_rate < all_ages_unemp_rate:
        recent_grads_lower_unemp_count += 1
    elif all_ages_unemp_rate < recent_grads_unemp_rate:
        all_ages_lower_unemp_count += 1
        
print(recent_grads_lower_unemp_count)
print(all_ages_lower_unemp_count)
```

#### Grouped Bar Charts

```python
# get the share of men and add it as a column
#recent_grads["ShareMen"] = recent_grads["Men"] / recent_grads["Total"]
# or
recent_grads["ShareMen"] = 1 - recent_grads["ShareWomen"]

# get only the art majors as a new data frame
art_grads = recent_grads[recent_grads["Major_category"] == "Arts"]

# create a figure instance
fig = plt.figure()
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(1,1,1)

# generate a range to account for majors
import numpy as np
locs = np.arange(art_grads.shape[0])

# create barplot of ShareMen only
bar1 = ax1.bar(left = locs, height = art_grads["ShareMen"].tolist(), width = 0.35)
ax1.set_xticklabels(art_grads["Major"].tolist(), rotation=90)

# add bars for ShareWomen
offset_locs = locs + 0.35
bar2 = ax1.bar(left = offset_locs, height = art_grads["ShareWomen"].tolist(), width = 0.35, color = "green")
ax1.set_xticks(offset_locs)

# add legend
# How to create a legend: http://matplotlib.org/users/legend_guide.html
plt.legend((bar1, bar2), ("ShareMen", "ShareWomen"), loc="upper left")

# show grid
plt.grid()
```

* Structure your Matplotlib code as a function so you can reuse the code. Write a function that takes in a DataFrame, takes in a list of column names, and generates a scatter matrix for combinations of columns.
    * While the scatter matrix you generated in this guided project used 2 columns, how can you generalize the code to handle n columns.
    * As n gets larger, how do you dynamically specify the figsize parameter when creating the Plot instance so the data visualization is legible with more subplots.


```python
def scatter_matrix(df, columns):
    n = len(columns)
    # create a figure instance
    fig = plt.figure(figsize=(4*n,4*n))

    # four subplots(axes)
    ax1 = fig.add_subplot(n,n,1)
    ax2 = fig.add_subplot(n,n,2)
    ax3 = fig.add_subplot(n,n,3)
    ax4 = fig.add_subplot(n,n,4)

    # counter for the subplot
    c = 0
    
    # plot in scatter matrix
    for i in range(n):
        for j in range(n):
            ax = fig.add_subplot(n,n,c+1)
            if i == j:
                ax.hist(df[columns[i]])
            else:
                ax.scatter(df[columns[i]], df[columns[j]])
            c += 1

    # remove spacing
    fig.subplots_adjust(wspace=0, hspace=0)

# example of function call with three or four columns
scatter_matrix(filtered_grads, ["ShareWomen", "Total", "Employed", "Unemployment_rate"])
    
```
